## Curse of Dimensionality
The Curse of Dimensionality refers to the challenges and problems that arise when dealing with high-dimensional data. It affects data analysis as follows:
 - As the number of dimensions increases, algorithms require more computational resources.
 - High-dimensional data can lead to overfitting in machine learning models.
 - It becomes harder to infer meaningful patterns with many dimensions.
 - Data points become more spread out, making it challenging to find neighbors or similarities.

One of the solutions to this curse is dimensionality reduction and this could be achieved by applying Principal Component Analysis (PCA) on dataset.

## Applying Principal Component Analysis (PCA) for Dimensionality Reduction.

### Importing necessary Libraries for PCA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

### Applying label encoding on dataset to convert categorical features (string) to numerical classes. This is done to incorporate the whole data for further calculations

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for i in df.columns:
    df[i] = le.fit_transform(df[i])
```

### Standardization (mean centering and scaling to unit variance). Standardization ensures that all features have equal influence in the PCA analysis, and it helps in capturing the true variance of each feature.

```python
scaler = StandardScaler()
data_standardized = scaler.fit_transform(df)
data_standardized
```

### Applying PCA to transform the original data into a new set of orthogonal features (principal components) that explain the most variance in the data.

```python
pca = PCA()
pca.fit(data_standardized)
```

### Calculating the explained variance ratio tells us the proportion of the total variance in the data that is explained by each principal component, helping to understand the significance of each component.

```python
explained_variance_ratio = pca.explained_variance_ratio_
explained_variance_ratio
```

### By calculating the cumulative explained variance, we can determine the number of components required to retain a certain percentage of the total variance. In this case, we aim to retain 99.5% (as mentioned in assignment) of the variance, which balances dimensionality reduction and information preservation.

```python
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
num_components = np.argmax(cumulative_explained_variance >= 0.995) + 1
num_components
```

### Applying PCA on selected most significant components can lead to improved model performance, faster training times, and easier data interpretation.

```python
pca = PCA(n_components=num_components)
data_reduced = pca.fit_transform(data_standardized)
data_reduced
```
